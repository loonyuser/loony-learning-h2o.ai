Duplicate rows removal if any. In our case,there were 5 duplicates

```python
package_data = package_data.drop_duplicates(columns = package_data.columns)

package_data.shape
```

```python
package_data[
    [
        'ProdTaken',
        'TypeofContact',
        'CityTier',
        'Occupation',
        'Gender',
        'ProductPitched',
        'MaritalStatus',
        'Passport',
        'OwnCar',
        'Designation'
    ]
].levels()
```

```python
num_categorical_cols = ['Passport', 'OwnCar', 'ProdTaken']

for col in num_categorical_cols:
    package_data[col] = package_data[col].asfactor()
```

```python
package_data['Passport'].isfactor()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

package_data_pd = package_data.as_data_frame(use_pandas = True)

sns.countplot( x = package_data_pd['ProdTaken'])
```

Splitting the data into Train/Test with Train having 80% and test and validation 20% each

```python
package_data_train, package_data_test = package_data.split_frame(ratios = [0.8])

package_data_train.shape, package_data_test.shape
```

Predictors and response vars are defined

```python
predictors = ['Age', 'TypeofContact', 'CityTier', 'DurationOfPitch', 
              'Occupation', 'Gender', 'NumberOfPersonVisiting',
              'NumberOfFollowups', 'ProductPitched', 'PreferredPropertyStar', 
              'MaritalStatus', 'NumberOfTrips', 'Passport', 'PitchSatisfactionScore',
              'OwnCar', 'NumberOfChildrenVisiting', 'Designation', 'MonthlyIncome']

response = 'ProdTaken'
```

Fitting a Generalised Linear estimator i.e Logistic Regression Model.
Note that 'standardize = True is by default i.e Standardize numeric columns to have zero mean and unit variance

```python
from h2o.estimators.glm import H2OGeneralizedLinearEstimator

package_glm = H2OGeneralizedLinearEstimator(family = 'binomial', standardize = True)

package_glm.train(predictors, response, training_frame = package_data_train, 
                  validation_frame = package_data_test)
```

Coefficients of predictors are obtained

```python
package_glm.coef()
```

```python
package_glm._model_json['output']['coefficients_table']
```

prediction on test set can be obtained( note that we are validating our model i.e validation set is also this test set). Prediction probabilities are added as column for 2 classes

```python
pred = package_glm.predict(package_data_test)

pred
```

p values will only be computed for coefficients if compute_p_values is set to true, and this will also only work if there is no regularisation, which requires lambda=0 and remove_collinear_columns = True.

```python
package_glm = H2OGeneralizedLinearEstimator(family = 'binomial', remove_collinear_columns = True,
                                            lambda_ = 0, compute_p_values = True)

package_glm.train(predictors, response, training_frame = package_data_train, 
                  validation_frame = package_data_test)
```

Note that p-values are there

```python
package_glm._model_json['output']['coefficients_table']
```

Metrics scores can be obtained . here threshold maximising the metric is used as threshold argument is None

```python
import pprint

print('\nAccuracy_score with thresholds :')
pprint.pprint(package_glm.accuracy(train = True, valid = True))

print('\nF1_score with thresholds :')
pprint.pprint(package_glm.F1(train = True, valid = True))

print('\nPrecision_score with thresholds :')
pprint.pprint(package_glm.precision(train = True, valid = True))

print('\nRecall with thresholds :')
pprint.pprint(package_glm.recall(train = True, valid = True))
```

```python
pred = package_glm.predict(package_data_test)

pred
```

Random forest model is fit

```python
from h2o.estimators import H2ORandomForestEstimator

package_rf_model = H2ORandomForestEstimator(binomial_double_trees = True)

package_rf_model.train(x = predictors, y = response, training_frame = package_data_train,
                       validation_frame = package_data_test)
```

```python
perf = package_rf_model.model_performance()

perf
```

Model performance can be obtained

Generating predictions on a validation set (if necessary):

```python
predictions = package_rf_model.predict(package_data_test)

predictions
```

Classes are imbalanced so we can balance the classes. In the output, it can be observed that both classes are around 3167.

```python
package_upsampled_rf_model = H2ORandomForestEstimator(balance_classes = True, binomial_double_trees = True)

package_upsampled_rf_model.train(x = predictors, y = response, 
                                 training_frame = package_data_train, 
                                 validation_frame = package_data_test)
```

```python
perf = package_upsampled_rf_model.model_performance()

perf
```

```python
print('\nAccuracy_score with thresholds :')
pprint.pprint(package_upsampled_rf_model.accuracy(train = True, valid = True))

print('\nF1_score with thresholds :')
pprint.pprint(package_upsampled_rf_model.F1(train = True, valid = True))

print('\nPrecision_score with thresholds :')
pprint.pprint(package_upsampled_rf_model.precision(train = True, valid = True))

print('\nRecall with thresholds :')
pprint.pprint(package_upsampled_rf_model.recall(train = True, valid = True))
```

```python
predictions = package_upsampled_rf_model.predict(package_data_test)

predictions
```
